Q1. What is Flask Framework? What are the advantages of Flask Framework?


*_* Flask is a micro web framework for Python that is designed to be lightweight and easy to use. It is often classified as a "micro" framework because it does not include specific tools or libraries for tasks like database abstraction, authentication, and form validation. Instead, Flask allows developers to choose the tools they want to use for these tasks, providing a flexible and modular structure.

Key features and advantages of Flask:-

* Lightweight and Simple: Flask is minimalistic and straightforward, making it easy to learn and use. It does not come with   unnecessary built-in features, giving developers the flexibility to choose the components they need for their specific       application.

* Modular Design: Flask follows the WSGI (Web Server Gateway Interface) standard, making it compatible with other WSGI-      compliant frameworks and servers. Its modular design allows developers to add or remove components as needed.

* Jinja2 Templating: Flask uses the Jinja2 template engine, which allows developers to easily generate dynamic HTML content. Templates support inheritance and macros, providing a powerful way to create reusable and maintainable views.

* Built-in Development Server: Flask comes with a built-in development server, making it easy to start developing and testing applications without the need for additional server configurations.

* Extensions: While Flask is minimalistic, it also supports a wide range of extensions that can be added to enhance functionality. These extensions cover areas such as authentication, database integration, form validation, and more

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, a popular web framework for Python, "app routing" refers to the process of defining URL patterns and associating them with specific functions or views in your application. This is typically done using the @app.route decorator. The routes define how URLs should be handled by our Flask application & How we Can Acess The our URL.

The main reasons for using app routes in Flask are:

* URL Mapping: App routes allow you to map specific URLs to corresponding functions or views in your application. This helps organize your code and makes it clear which part of your code is responsible for handling which URLs.

* Modularity: By associating specific functions with specific routes, Flask promotes a modular and organized code structure. Each route and its corresponding function can be designed to handle a specific aspect of your application.

* Readability: Using app routes and decorators makes the code more readable and expressive. Developers can easily understand which function corresponds to which URL endpoint.

* Dynamic URLs: Flask supports dynamic URL components, allowing you to define routes with variable parts. This enables you to capture values from the URL and use them in your view functions.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

ANSWER => In Flask, the url_for() function is used for URL building. It generates a URL for the given endpoint with the method provided. It takes the endpoint name as its first argument and generates a URL for that endpoint. Additionally, we can pass values for the variables in the URL pattern, and url_for() will incorporate them into the URL.

let us Taking an Example of makeing the URL :-

```python
from flask import Flask, url_for
```

```python
app = Flask(__name__)

@app.route('/')
def index():
    return "Hello this is the my Home_Page!"
@app.route('/user/<username>')
def user_profile(username):
    return f'Hello, {username}!'
@app.route('/dashboard')
def dashboard():
    user_url = url_for('user_profile' , username='shubham_shah' )
    return f'This is the dashboard! check out the user profile:{user_url}'
if __name__=="__main__":
    app.run(host="0.0.0.0")
```
